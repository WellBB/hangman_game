```python
# Development of a hangman game

import numpy as np

print('>>>>> Hangman V1.0 <<<<<\n')

print('''
 +====+
 |    |
      |
      |
      |
      |
===========\n''')

class gallows():
    def mistakes(mistakes = 0):
        if mistakes == 0:
            print('''
 +====+
 |    |
      |
      |
      |
      |
===========\n''')
        elif mistakes == 1:
            print('''
 +====+
 |    |
 O    |
      |
      |
      |
===========\n''')
        elif mistakes == 2:
            print('''
 +====+
 |    |
 O    |
 |    |
      |
      |
===========\n''')
        elif mistakes == 3:
            print('''
 +====+
 |    |
 O    |
/|    |
      |
      |
===========\n''')
        elif mistakes == 4:
            print('''
 +====+
 |    |
 O    |
/|\   |
      |
      |
===========\n''')
        elif mistakes == 5:
            print('''
 +====+
 |    |
 O    |
/|\   |
/     |
      |
===========\n''')
        elif mistakes == 6:
            print('''
 +====+
 |    |
 O    |
/|\   |
/ \   |
      |
===========\n
You lost! Thanks for playing!''')

with open('wordFile.txt', 'r') as wordFile:
    fullData = wordFile.read().split('\n')
randomChoice = np.random.choice(a = fullData, size = 1)
print('\nYour word is:','_ ' * len(randomChoice[0]))

letter = []
rightLetters = []
wrongLetters = []
validLetters = ['a', 'b', 'c', 'd', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
countMistakes = 0

while len(list(filter(lambda x: list(randomChoice[0])[x] in rightLetters, range(len(randomChoice[0]))))) != len(randomChoice[0]):
    letter.clear()
    letter.append(input('\nEnter a valid letter:'))
    while not (letter[0] in validLetters) == True and len(list(letter)) == 1 and letter == None:
        print('\nInvalid option!')
        letter.clear()
        letter.append(input('\nPlease enter a valid or not yet entered letter:'))
    if letter[0] in list(randomChoice[0]):
        rightLetters.append(letter[0])
    else:
        wrongLetters.append(letter[0])
        countMistakes = countMistakes + 1
    gallows.mistakes(mistakes = countMistakes)
    controlString = ''
    for y in list(randomChoice[0]):
        if (y in rightLetters) == True:
            controlString += '%s ' %(y)
        else:
            controlString += '_ '
    print('\nYour word is:', controlString)
    print('\nRight Letters:', ', '.join(map(str, rightLetters)))
    print('\nWrong Letters:', ', '.join(map(str, wrongLetters)))
else:
    print('\nYou win! Thanks for playing!')
```
